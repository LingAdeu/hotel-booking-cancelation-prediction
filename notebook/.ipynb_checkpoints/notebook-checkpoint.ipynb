# <b>1 Introduction</b>

<b>Problem Statement</b>: In the hospitality industry, managing hotel bookings efficiently is crucial for maximizing revenue and optimizing resources. However, one of the significant challenges faced by hotel managers is room booking cancellations, which can lead to revenue loss and operational inefficiencies. To address this challenge, my aim is to develop a machine learning classification model to predict hotel booking cancellations.

<b>Objective</b>: The objective of this project is to predict hotel booking cancellations identify and mitigate high-risk bookings. By leveraging historical booking data and various features associated with each booking, I intend to build a predictive model that can classify bookings into "canceled" or "not canceled" categories. This model will enable hotel managers to allocate resources effectively, optimize inventory management by reducing potential revenue loss due to cancellations.

<b>Analytical Approach</b>
- <b>Step 1</b>: Perform an exploratory data analysis (EDA) to understand the customer behavior who canceled the hotel bookings.
- <b>Step 2</b>: Build a classification model to predict the probability of customers canceling the hotel booking.
- <b>Step 3</b>: Identify what factors underlie the cancelation of a hotel booking.

<b>Metrics</b>: The focus of this analysis is on customers who canceled their hotel room bookings. 

| Error Type | Description | Busines Impacts |
| :--- | :--- | :--- |
| <b>False Positive (FP)</b> or Type I error | The model incorrectly indentifies booking as canceled when it is not. | Lead to overbooking or an unnecessary action taken by the hotel, e.g., allocating additional resources to manage a cancellation that never occurs. This can result in operational inefficiencies and potential loss in revenue. |
| <b>False Negative (FN)</b> or Type II error | The model incorrectly predicts that a hotel booking will not be canceled but in reality, it is canceled. In other words, the model misidentifies a canceled booking as a non-canceled one. | Hotel being unprepared for cancellations, leading to issues such as unavailability of rooms for other potential guests or excess staff scheduled to handle the expected bookings. |

- <span style='color:red'>Simulation of Type I Error</span>:
	- Financial implication: The hotel will need to face a either compensation cost for providing an alternative room or rebooking cost in another hotel. 
		- Primary compensation cost: The hotel should compensate the guests by providing a similar or better room type. While the hotel may pay \$0 cost if there is another similar room, the hotel will lose between \$41 and \$90 to upgrade the hotel room to normal or luxury type. 
			- **Reference**: Budget (\$58 per night), mid-range hotel (USD99), and luxury (USD89) ([Budget Your Trip](https://www.budgetyourtrip.com/hotels/portugal-PT)).
	- Non-financial implication: Incorrect cancellation can harm hotel's reputation because guests can leave negative reviews which will deter potential guests from making a booking.
- <span style='color:red'>Simulation of Type II Error</span>: 
	- Financial implication: The average price of hotel room rent is USD99/night in Portugal. To put this standard pricing into context, whenever the model gives a false negative, the hotel will lose USD99 per room. 
		- **Reference:** [Budget Your Trip](https://www.budgetyourtrip.com/hotels/portugal-PT)
	- Non-financial implication: Lorem ipsum

- `required_car_parking_space` (integer): Number of car parking spaces required by the customer.
- `total_of_special_request` (integer): Number of special requests made by the customer (e.g. twin bed or high floor).

<b>Target</b>
- `is_canceled` (integer): Value indicating if the booking was canceled (1) or not (0).

#### <b>Outlier Identification & Treatment</b>

```python
# create function to detect outliers using IQR
def detect_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df.loc[(df[column] < lower_bound) | (df[column] > upper_bound)]
    print(f"The number of outliers in '{column}' is {len(outliers)} or {(len(outliers)/len(df)*100):.2f}% of the total data ({len(df)} rows).")
```

```python
# detect outliers
for col in df.select_dtypes(include=np.number).columns:
    detect_outliers(df, col)
```

```python
from sklearn.model_selection import train_test_split

RANDOM_STATE = 42

# Split the dataset into training and testing sets
df_seen, df_unseen = train_test_split(df, stratify=df['is_canceled'], test_size=0.2, random_state=RANDOM_STATE)
```

```python
from pycaret.classification import *
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import KBinsDiscretizer
from sklearn.preprocessing import OneHotEncoder
from pycaret.internal.preprocess.transformers import TransformerWrapper
from category_encoders import BinaryEncoder

pipeline = [
    ('Equal frequency binning', TransformerWrapper(
        include=['previous_cancellations', 'booking_changes', 
                'days_in_waiting_list', 'required_car_parking_spaces', 
                'total_of_special_requests'], 
        transformer=KBinsDiscretizer(strategy='quantile'))),
    ('imputer', TransformerWrapper(
        include=['country'], 
        transformer=SimpleImputer (strategy='most_frequent'))),
    ("onehot", TransformerWrapper(
        include=['market_segment', 'deposit_type', 
                 'customer_type', 'reserved_room_type '], 
        transformer=OneHotEncoder())),
    ("binary", TransformerWrapper(
        include=['country'], 
        transformer=BinaryEncoder()))
]
```

<b>Comment</b>
- <b>Feature transformation</b>
    - <b>Equal frequency discretization</b>: Because there are outliers, column `previous_cancellations`, `booking_changes`,  `days_in_waiting_list`, `required_car_parking_spaces`, `total_of_special_requests` are discretized using equal frequency binning. This technique splits the data based on quantile into the same frequencies.
    - <b>Onehot encoding</b>: `market_segment`, `deposit_type`, `customer_type`, `reserved_room_type`
    - <b>Binnary encoding</b>: As there are more than 15 categories in `countries`, onehot encoding is not recommended. Here, binary encoding is applied.
- <b>Data imputation</b>: The missing values (0.42%) in column `country` are inputed by using most frequent value.
